## Feature Engineering

```python
df.columns
```

```python
X = df[['Average Cost for two', 'Has Table booking', 'Has Online delivery', 'Price range']]
```

```python
# above 2 below 2.5  = poor
# above 2.5 below 3.5 = average
# above 3.5 below 4.0 = good
# above 4.0 below 4.5 = very good
# above 4.5 = excellent
```

```python
y = df[["Aggregate rating"]]
y
```

```python
X
```

```python
df['Price range'].value_counts()
```

```python
from sklearn import preprocessing
```

```python
labelencoder = preprocessing.LabelEncoder()
```

```python
X['Has Table booking'] = labelencoder.fit_transform(X["Has Table booking"])
```

```python

X
```

```python
labelencoder.classes_ #index 0 is NO and index 1 is YES
```

```python
X['Has Online delivery'] = labelencoder.fit_transform(df["Has Online delivery"])
```

```python
X
```

```python
labelencoder.classes_
#has online delivery 0 is no and 1 is yes
#Has table booking 0 is NO and 1 is YES
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
```

```python
def modelresult(predictions):
    print("Mean absolute error of model is {}".format(mean_absolute_error(y_test, predictions)))
    print("Mean absolute error of model is {}".format(np.sqrt(mean_squared_error(y_test, predictions))))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
predslr = lr.predict(X_test)
```

```python
modelresult(predslr)
```

```python
predslr
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
```

```python
svrmodel = SVR()
```

```python
param_gridsvr = {"C":[ 0.1, 0.5], "kernel":["linear","poly"], "degree":[2]}
```

```python
gridsvr  = GridSearchCV(svrmodel, param_gridsvr)
```

```python
gridsvr.fit(X_train, y_train)
```

```python
predgridsvr = gridsvr.predict(X_test)
```

```python
modelresult(predgridsvr)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
param_grid = {
    "max_depth":[10,30],
    "min_samples_leaf":[1,2],
    "min_samples_split":[2,5]
}
```

```python
treemodel = DecisionTreeRegressor()
```

```python
grid_tree = GridSearchCV(estimator=treemodel, param_grid=param_grid)
```

```python
grid_tree.fit(X_train, y_train)
```

```python
treepredictions = grid_tree.predict(X_test)
```

```python
modelresult(treepredictions)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfmodel = RandomForestRegressor()
```

```python
param_gridfr = {"max_depth":[5,15], "n_estimators":[2,5,10] }
```

```python
gridrfr = GridSearchCV(rfmodel, param_gridfr)
```

```python
gridrfr.fit(X_train, y_train)
```

```python
randomforestpredictions = gridrfr.predict(X_test)
modelresult(randomforestpredictions)
```

```python
gridrfr.best_params_
```

```python
grid_tree.best_params_
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor()
knn_param_grid = {"n_neighbors":[9,11,13,14, 15, 16]}
knn_grid_search = GridSearchCV(knn, knn_param_grid)
```

```python
knn_grid_search.fit(X_train, y_train)
```

```python
knnpreds = knn_grid_search.predict(X_test)
```

```python
modelresult(knnpreds)
```

```python
knn_grid_search.best_params_
```

```python
from sklearn.ensemble import AdaBoostRegressor
ada = AdaBoostRegressor()

ada_param_grid = {"n_estimators":[50,100,200],
                      "learning_rate":[0.1,0.5,1]}

ada_grid_search = GridSearchCV(ada, ada_param_grid)
```

```python
ada_grid_search.fit(X_train, y_train)
```

```python
adapreds = ada_grid_search.predict(X_test)
```

```python
modelresult(adapreds)
```
